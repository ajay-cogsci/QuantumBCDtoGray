```
from qiskit import Aer, QuantumCircuit, execute
from qiskit.visualization import plot_histogram
```

```
bcd2gray = QuantumCircuit(4,4)

############ ENCODE#################################
# Encode as required by uncommenting required statement. Current input encoded is - 0101. You can see the result as - 0111

bcd2gray.x(0)
#bcd2gray.x(1)
bcd2gray.x(2)
#bcd2gray.x(3)
bcd2gray.barrier()
#####################################################

bcd2gray.cx(1,0)
bcd2gray.measure(0,0)

bcd2gray.cx(2,1)
bcd2gray.measure(1,1)

bcd2gray.cx(3,2)
bcd2gray.measure(2,2)

bcd2gray.measure(3,3)


bcd2gray.draw()
```

```
counts = execute(bcd2gray, Aer.get_backend("qasm_simulator")).result().get_counts()
plot_histogram(counts)
```
